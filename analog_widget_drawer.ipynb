```python
import ipywidgets as widgets
from ipywidgets import interactive,GridBox,Dropdown,Layout
from IPython.display import display
```

```python
file = open("white.png", "rb")
im_none = file.read()
file = open("gnd.png", "rb")
im_gnd = file.read()
file = open("nmos.png", "rb")
im_nmos = file.read()
file = open("pmos.png", "rb")
im_pmos = file.read()
```

```python
wg_none = widgets.Image(value=im_none,format='png',width=50,height=50,)
wg_gnd = widgets.Image(value=im_gnd,format='png',width=50,height=50,)
wg_nmos = widgets.Image(value=im_nmos,format='png',width=50,height=50,)
wg_pmos = widgets.Image(value=im_pmos,format='png',width=50,height=50,)
```

```python
def im_select(name):
  if(name == ''):
    display(wg_none)
  if(name == 'gnd'):
    display(wg_gnd)
  if(name == 'pmos'):
    display(wg_pmos)
  if(name == 'nmos'):
    display(wg_nmos)
```

```python
option_widget = widgets.Dropdown(options=['','gnd','pmos','nmos'],description=' ',layout={'width': 'max-content'})
f1 = interactive(im_select, name=option_widget);
display(f1)
```

```python
selection = GridBox(children=[interactive(im_select, name=Dropdown(options=['','gnd','pmos','nmos'],description=' ',layout={'width': 'max-content'})) for i in range(9)],
                    layout=Layout(
                      width='100%',
                      grid_template_columns='200px 200px 200px',
                      grid_template_rows='120px 120px 120px')
                    )
```

```python
tab_nest = widgets.Tab()
tab_nest.set_title(0, 'Easy Circuit Drawer')
tab_nest.children = [selection]
display(tab_nest)
```
